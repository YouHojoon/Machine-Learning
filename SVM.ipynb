```
import numpy as np
import matplotlib.pyplot as plt

def GaussianKernel(v1,v2,sigma):
  res = np.exp(-np.sum(np.power(v1-v2,2))/2*np.power(sigma,2))
  return res
```

```
val = GaussianKernel(np.array([0,0]),np.array([3,5]),1.0)
print(val)
```

```
d = 100
x1 = np.linspace(0,6,d) #0~6까지 100개의 마크를 세움
x2 = np.linspace(0,10,d)
l1 = [3,5]

Z = np.zeros((d,d))
sigma = 1.5
for i,t1 in enumerate(x1):
  for j,t2 in enumerate(x2):
    val = GaussianKernel(np.array([t1,t2]), l1, sigma)
    Z[i,j] = val
```

```
from mpl_toolkits import mplot3d

pX, pY = np.meshgrid(x1,x2)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(pX, pY, Z, 100, cmap='jet')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('Z')
plt.show()
```

```
import io
from google.colab import files
data_file = files.upload()
```

```
from scipy.io import loadmat
data=loadmat(io.BytesIO(data_file['data1.mat']))
print(data)

X = data['X']
y = data['y']
m= len(y)

print(X.shape,y.shape)
```

```
pos = []
neg = []

for i, val in enumerate(y):
  if val == 1:
    pos.append(i)
  else:
    neg.append(i)
print(pos)
print(neg)
```

```
plt.plot(X[pos,0].reshape(-1),X[pos,1].reshape(-1),'b+',label='y=1')
plt.plot(X[neg,0].reshape(-1),X[neg,1].reshape(-1),'ro',label='y=0')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc='upper right')
plt.title('data 1')
plt.show()
```

```
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC

svm_clf = Pipeline([
                    ('scaler',StandardScaler()),
                    ('linear_svc', LinearSVC(C=100, loss='hinge'))
])
svm_clf.fit(X,y)
```

```
svm_clf.predict([[2.3,2.7],[3.7,3.9]])
```

```
x_min, x_max = X[:,0].min()-0.2, X[:,0].max()+0.2
y_min, y_max = X[:,1].min()-0.2, X[:,1].max()+0.2
h=.02
xx,yy=np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
Z = svm_clf.predict(np.c_[xx.ravel(),yy.ravel()])

Z=Z.reshape(xx.shape)
plt.figure(1)
plt.pcolormesh(xx,yy,Z,cmap=plt.cm.Paired)

plt.plot(X[pos,0].reshape(-1),X[pos,1].reshape(-1),'b+',label='y=1')
plt.plot(X[neg,0].reshape(-1),X[neg,1].reshape(-1),'ro',label='y=0')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc='upper right')

plt.xlim(xx.min(),xx.max())
plt.ylim(yy.min(),yy.max())
plt.xticks(())
plt.yticks(())
plt.title('data 1')
plt.show()
```

```
data = loadmat(io.BytesIO(data_file['data2.mat']))
np.set_printoptions(edgeitems=2,threshold=50)
print(data)

X=data['X']
y=data['y']
m=len(y)

print(X.shape,y.shape)
```

```
pos=[]
neg=[]

for i,val in enumerate(y):
  if val==1:
    pos.append(i)
  else:
    neg.append(i)

print(pos)
print(neg)
```

```
plt.plot(X[pos,0].reshape(-1),X[pos,1].reshape(-1),'b+',label='y=1')
plt.plot(X[neg,0].reshape(-1),X[neg,1].reshape(-1),'ro',label='y=0')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc='upper right')
plt.title('data2')
plt.show()
```

```
from sklearn.svm import SVC
svm_clf = Pipeline([
                    ('scaler',StandardScaler()),
                    ('nonliner_svc',SVC(kernel='rbf',gamma=10, C=1))
                    #gamma = 1/sigma
                    #rbf = gaussian kernel
                    #poly = polynomial kernal
])
svm_clf.fit(X,y)
```

```
x_min, x_max = X[:,0].min()-0.2, X[:,0].max()+0.2
y_min, y_max = X[:,1].min()-0.2, X[:,1].max()+0.2
h=.02
xx,yy=np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
Z = svm_clf.predict(np.c_[xx.ravel(),yy.ravel()])

Z=Z.reshape(xx.shape)
plt.figure(1)
plt.pcolormesh(xx,yy,Z,cmap=plt.cm.Paired)

plt.plot(X[pos,0].reshape(-1),X[pos,1].reshape(-1),'b+',label='y=1')
plt.plot(X[neg,0].reshape(-1),X[neg,1].reshape(-1),'ro',label='y=0')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc='upper right')

plt.xlim(xx.min(),xx.max())
plt.ylim(yy.min(),yy.max())
plt.xticks(())
plt.yticks(())
plt.title('data 2')
plt.show()
```
